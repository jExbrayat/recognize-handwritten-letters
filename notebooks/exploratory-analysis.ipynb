```python
plt.figure(figsize=(10, 5))
letter_col = 0
letter_counts = df.iloc[:, letter_col].value_counts().sort_index()

plt.xlabel("Lettres")
plt.ylabel("Nombre d'occurrences")
plt.title("Distribution des lettres dans le dataset")
plt.xticks(
    ticks=range(26),
    labels=letters,
)
plt.bar(letter_counts.index, letter_counts.values)
plt.show()
```
