```python
new_data = np.array(img).reshape((-1, 784))
new_data.shape
```

```python
def convert_int_to_letter(int_list: list[int]) -> list[str]:
    return list(map(lambda i: letters_list[i], int_list))
```

```python
def run_inference(model, data):
    processed_data = preprocess(data)

    input_name = model.get_inputs()[0].name
    output_name = model.get_outputs()[0].name

    result = model.run([output_name], {input_name: processed_data})

    return convert_int_to_letter(result[0])
```

```python
run_inference(session, new_data)
```

```python
run_inference(session, x)
```
